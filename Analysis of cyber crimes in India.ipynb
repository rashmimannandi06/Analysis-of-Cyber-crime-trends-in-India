```python
'''
We have the statistics for crimes for the individual year. We can also find the state which is worst affected for a particular crime
over the 3 years, by taking the average for a particular category
'''


maximums = []
maxstates = []

n1,n2,n3=0,0,0
states = list(data1.STATES)

#for each category we want to find the state with the highest average across the 3 years
for i in range(1, len(data1.columns)):
    n1 = data1[data1.columns[i]]
    n2 = data2[data1.columns[i]]
    n3 = data3[data1.columns[i]]
    average = (n1+n2+n3)/3
    #for the state:
    for j in range(len(average)):
        if (average[j] == max(average) and max(average)!=0):
            maximums.append(max(average))
            maxstates.append(data1.STATES[j])

#making a dataframe of the data
df4 = pd.DataFrame({'Categories':categories, 'STATES':maxstates, 'Highest Average':maximums})
df4
```

```python
#we want to present the above data cleanly into one single table of values, where:
#rows = category of crime
#columns = years(2018,2019,2020), state name for each year and then state with highest average and the exact number

#creating the required final dataframe

final_df = pd.DataFrame()

#creating the columns
final_df['Categories'], final_df['2018'], final_df['state_2018'], final_df['2019'], final_df['state_2019'], final_df['2020'], final_df['state_2020'] =  categories, df1.Figures, df1.States, df2.Figures, df2.States, df3.Figures, df3.States
#adding the average column
final_df["Highest Average"], final_df["State"] = df4['Highest Average'], df4.STATES
final_df.drop([10,20,21], axis=0)
```
